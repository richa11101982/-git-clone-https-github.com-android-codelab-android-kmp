Here's an example of a basic Cartopy plot showing a world map with coastlines and country borders.

```
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np # Import numpy for generating random data

# Create a figure and an axes object with a Plate Carree projection
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Add land and ocean features with colors
ax.add_feature(cfeature.LAND, color='lightgray')
ax.add_feature(cfeature.OCEAN, color='lightblue')

# Add coastlines and country borders
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='gray')

# Set the extent of the map (optional, shows the whole world by default)
ax.set_global()

# Add gridlines
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

# --- New code to plot data points ---
# Generate some random data points (latitude and longitude)
np.random.seed(42) # for reproducibility
num_points = 50
random_lons = np.random.uniform(-180, 180, num_points)
random_lats = np.random.uniform(-90, 90, num_points)

# Plot the data points
# The 'transform=ccrs.PlateCarree()' argument is crucial here. It tells cartopy
# that the random_lons and random_lats are in Plate Carree coordinates (standard lat/lon)
# and it will transform them correctly to the map's projection.
ax.scatter(random_lons, random_lats, color='red', marker='o', s=50, alpha=0.7,
           transform=ccrs.PlateCarree(), label='Random Data Points')

# Add a legend to explain the data points
ax.legend()
# --- End of new code ---

# Add a title
plt.title('Cartopy World Map with Land, Ocean, and Data Points')

# Display the plot
plt.show()
```

```
import pandas as pd

# Load the california_housing_train.csv file into a pandas DataFrame
housing_df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Display the first 5 rows of the DataFrame
print('First 5 rows of the California Housing DataFrame:')
display(housing_df.head())
```

```
print('Summary statistics for housing_df:')
display(housing_df.describe())
```

To visualize the relationship between `median_income` and `median_house_value`, we'll create a scatter plot. This will allow us to observe if there's a correlation between a district's median income and the median value of its houses.

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=housing_df, alpha=0.6)
plt.title('Relationship between Median Income and Median House Value')
plt.xlabel('Median Income (in tens of thousands USD)')
plt.ylabel('Median House Value (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Now, let's visualize the geographical distribution of housing in California, coloring the points by `median_house_value`. This will give us an idea of how housing prices vary across different regions.

```
import matplotlib.pyplot as plt

# Create a scatter plot of latitude vs longitude, colored by median_house_value
plt.figure(figsize=(10, 8))
scatter = plt.scatter(housing_df['longitude'], housing_df['latitude'],
                      c=housing_df['median_house_value'], cmap='viridis', s=housing_df['population']/100,
                      alpha=0.4, edgecolors='w', linewidths=0.5)

plt.colorbar(scatter, label='Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('California Housing Prices by Location')
plt.grid(True, alpha=0.3)
plt.show()
```
